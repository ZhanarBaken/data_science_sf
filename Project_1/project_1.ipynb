3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#прописываем функцию, которая из признака 'Опыт работы' вычислит количество месяцев опыта работы 
def get_experience(arg):
    if arg is np.nan or arg == 'Не указано':
        return None #возвращаем NaN, если опыт работы не указан или пусто 
    year_key_words=['год', 'года', 'лет'] #создам список всех вариаций слова "год"
    month_key_words=['месяц', 'месяца', 'месяцев'] #создам список всех вариаций "месяц"
    args_splited = arg.split(' ')[:7] #вырезаем слова с первого по седьмое
    year = 0 
    month = 0 #создаем счетчики к годам и месяцам 
    for i in range(len(args_splited)): 
        if args_splited[i] in month_key_words:
            month = args_splited[i-1] #через цикл вычислем количество месяцев 
        if args_splited[i] in year_key_words:
            year = args_splited[i-1] #через цикл вычислем количество лет
    return int(year)*12 + int(month) #переводим в целочислительный тип и прибавляем 


#применяем эту функцию и сохраняем данные в новый признак 'Опыт работы (месяц)'
data['Опыт работы (месяц)'] = data['Опыт работы'].apply(get_experience)



#вычисляем медиану опыта работы в месяцах
ex_median = round(data['Опыт работы (месяц)'].median())
print(f'Медианный опыт работы {ex_median} месяцев') #выводим количество


#удаляем признак: 'Опыт работы'
data = data.drop('Опыт работы', axis=1)
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#прописываем функцию которая преоброзует признак 'Город, переезд, командировки' в 4 категории городов 
def get_category(arg):
    arg_split = arg.split(' ,')
    moscow = ['Москва']
    spb = ['Санкт-Петербург']
    million = ['город-миллионник']
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград']
    for x in arg_split:
        if x in moscow:
            return 'Москва'
        if x in spb:
            return 'Санкт-Петербург'
        if x in million_cities:
            return 'город-миллионник'
        else:
            return 'другие'
    
#применяем данную функцию     
data['Город'] = data['Город, переезд, командировки'].apply(get_category)
#проверяем, должно быть 4 уникальных значений
data['Город'].nunique() #4


#прописываем функцию, которая преоброзует готовых и не готовых к переезду на True и False
def get_ready_for_relocate(arg):
    if ('не готов к переезду' in arg) or ('не готова к переезду' in arg): #находим не готовых 
        return False #возвращаем для них False
    else: 
        return True #для всех остальных возвращаем True
    
#применяем данную функцию к признаку 'Город, переезд, командировки' и сохраняем в новый созданный признак 'Готовность к переезду'
data['Готовность к переезду'] = data['Город, переезд, командировки'].apply(get_ready_for_relocate)

#проверяем, должно быть 2 уникальных значений
data['Готовность к переезду'].nunique() #2


#прописываем функцию, которая преоброзует готовых и не готовых к командировкам на True и False
def get_ready_for_bisiness_trips(arg):
    if ('командировка' in arg):
        if ('не готов к командировкам' in arg) or ('не готова к командировкам' in arg): #находим не готовых 
            return False #возвращаем для них False
        else: 
            return True #для всех остальных возвращаем True
        
#применяем данную функцию к признаку 'Город, переезд, командировки' и сохраняем в новый созданный признак 'Готовность к командировкам'  
data['Готовность к командировкам'] = data['Город, переезд, командировки'].apply(get_ready_for_bisiness_trips)

#проверяем, должно быть 2 уникальных значений
data['Готовность к командировкам'].nunique() #2


#считаем сколько процентов соискателей живут в Санкт-Петербурге, округляем до целого
saint_petersburg = round((data[data['Город']=='Санкт-Петербург'].shape[0]/data.shape[0])*100)
print(f'В Санкт-Петербурге живут {saint_petersburg}% процентов соискателей') #выводим количество

#считаем сколько процентов соискателей готовы одновременно и к переездам, и к командировкам
mask1 = data['Готовность к переезду'] == True #создаем маску готовых к перрезду
mask2 = data['Готовность к командировкам'] == True #создаем маску готовых к командировкам 
#делим количество готовых на общее количество и окргуялем до целого умножив на 100
super_ready = round((data[mask1&mask2].shape[0]/data.shape[0])*100)
print(f'{super_ready}% соискателей готовы одновременно и к переездам, и к командировкам') #выводим количество


#удаляем признак: 'Город, переезд, командировки'
data = data.drop('Город, переезд, командировки', axis=1)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#создаем признаки-мигалки из признаков 'Занятость' и 'График' через лямбду 
data['полная занятость'] = data['Занятость'].apply(lambda x: True if 'полная занятость' in x else False)
data['частичная занятость'] = data['Занятость'].apply(lambda x: True if 'частичная занятость' in x else False)
data['проектная работа'] = data['Занятость'].apply(lambda x: True if 'проектная работа' in x else False)
data['волонтерство'] = data['Занятость'].apply(lambda x: True if 'волонтерство' in x else False)
data['полный день'] = data['График'].apply(lambda x: True if 'полный день' in x else False)
data['удалённая работа'] = data['График'].apply(lambda x: True if 'удалённая работа' in x else False)
data['вахтовый метод'] = data['График'].apply(lambda x: True if 'вахтовый метод' in x else False)
data['гибкий график'] = data['График'].apply(lambda x: True if 'гибкий график' in x else False)


#сичтаем сколько людей ищут проектную работу и волонтёрство, через фильтрацию 
freelancers = data[(data['проектная работа']==True) & (data['волонтерство']==True)].shape[0]
print(f'{freelancers} соискателей ищут проектную работу и волонтёрство') #выводим количество

#считаем сколько людей хотят работать вахтовым методом и с гибким графиком, через фильтрацию 
shifts = data[(data['вахтовый метод']==True)&(data['гибкий график'])].shape[0]
print(f'{shifts} соискателей хотят работать вахтовым методом и с гибким графиком') #выводим количество


#удаляем признаки: 'Занятость', 'График'
data = data.drop(['Занятость', 'График'], axis=1)
```

Нам бы хотелось видеть заработную плату в единой валюте, например, в рублях. Возникает вопрос, а где взять курс валют по отношению к рублю?

На самом деле язык Python имеет в арсенале огромное количество возможностей получения данной информации, от обращения к API Центробанка, до использования специальных библиотек, например pycbrf. Однако, это не тема нашего проекта.

Поэтому мы пойдем в лоб: обратимся к специальным интернет-ресурсам для получения данных о курсе в виде текстовых файлов. Например, MDF.RU, данный ресурс позволяет удобно экспортировать данные о курсах различных валют и акций за указанные периоды в виде csv файлов. Мы уже сделали выгрузку курсов валют, которые встречаются в наших данных за период с 29.12.2017 по 05.12.2019. Скачать ее вы можете **на платформе**

Создайте новый DataFrame из полученного файла. В полученной таблице нас будут интересовать столбцы:
* "currency" - наименование валюты в ISO кодировке,
* "date" - дата, 
* "proportion" - пропорция, 
* "close" - цена закрытия (последний зафиксированный курс валюты на указанный день).


Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Пропорция - это число, за сколько единиц валюты указан курс в таблице с курсами. Например, для казахстанского тенге курс на 20.08.2019 составляет 17.197 руб. за 100 тенге, тогда итоговый курс равен - 17.197 / 100 = 0.17197 руб за 1 тенге.

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#читаем новую тублицу с курсами валют 
ex = pd.read_csv('data/ExchangeRates.csv', sep=',')

#преоброзовываем столбец 'Обновление резюме' в формат datetime и выделяем дату
data['Обновление резюме'] = pd.to_datetime(data['Обновление резюме'], dayfirst=True).dt.date

#преоброзовываем столбец 'date' в формат datetime из новой таблицы с курсами 
ex['date'] = pd.to_datetime(ex['date'], dayfirst=True)
#создаем новый признак в основнойтаблице с таким же названием как в таблице с курсами
data['date'] = data['Обновление резюме']  
data['date'] = pd.to_datetime(data['date'])#тоже переводим в формат datetime


#функция которая вернет целочисленные цифры из зарплаты
def extraction_numbers(arg):
    extraction_list = [
        'руб.','грн.', 'USD', 'EUR',  'KGS', 'сум', 'AZN', 'KZT', 'бел.руб.'
        ] #создаем список слов, которые должны быть извлечены
    arg_splitted = arg.split(' ')
    new_arg = '' #создаем пустую строку, куда будем складывать цифры 
    for symb in arg_splitted: 
        if symb in extraction_list:
            symb = '' #если символ из списка, удаляем
        new_arg = new_arg + symb #остальыне складывам в пустую строку 
    return int(new_arg) #возвращаем строку в формате целочислного типа 

#создаем временный  признак 'ЗП(цифры)' только с цифрами, применив данную функцию
data['ЗП(цифры)'] = data['ЗП'].apply(extraction_numbers)


#функция которая извлекает валюты 
def extraction_currency(arg):
    extraction_list = [
        'руб.','грн.', 'USD', 'EUR',  'KGS', 'сум', 'AZN', 'KZT', 'бел.руб.'
        ] #используем тот же список 
    arg_splitted = arg.split(' ')
    new_arg = '' #создаем новую строку 
    for symb in arg_splitted:
        if symb in extraction_list:
            new_arg = new_arg + symb #если символ в списке, то наоброт добавляем в новую строку 
        else:
            symb ='' #если символа нету в списке, то удаляем
    return new_arg #возвращаем строку 

#применив данную функцию, создаем временный признак 'currency' с валютами 
data['currency'] = data['ЗП'].apply(extraction_currency)

#пишем функцию, которая переводит валюты в международный стандарт
def changing_to_iso(arg):
    if 'грн.' in arg:
        return 'UAH'
    elif 'бел.руб.' in arg:
        return 'BYN'
    elif 'сум' in arg:
        return 'UZS'
    else:
        return arg
#переводим валюты в международный стандарт в признаке data['currency']
data['currency'] = data['currency'].apply(changing_to_iso)


#объединяем две таблицы по дате и валюте 
merged = data.merge(
    ex, #датафрейм
    on=['date', 'currency'], #столбцы по которым идет объединение
    how='left' #тип объединения 
)

#заполняем пустые строки курса и пропорции единицей для рублей 
merged['close'] = merged['close'].fillna(1)
merged['proportion'] = merged['proportion'].fillna(1)

#создаем новый признак зарплаты в рублях 
merged['ЗП (руб)'] = (merged['ЗП(цифры)'] * merged['close'])/merged['proportion']

#считаем желаемую медианную заработную плату соискателей в таблице (в рублях)
salary_median =  round(merged['ЗП (руб)'].median())
print(f'{salary_median} рублей желаемая медианная заработная плата соискателей в таблице') #выводим количество

#удаляем не нужные признаки, из всех новых созданных, нам нужен только признак зарплаты в рублях
columns_for_daleting = ['ЗП(цифры)', 'currency', 'per', 'time',	'close', 'vol',	'proportion', 'ЗП', 'date'] 
merged = merged.drop(columns_for_daleting, axis=1)
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
#строим графики на одной плоскости 
fig, axes = plt.subplots(1, 2, figsize=(15, 4))

#гистограмма исходного признака
histplot = sns.histplot(data_cleaned['Возраст'], bins=30, ax=axes[0])
histplot.set_title('Распредление возраста соискателей');  #заголовок 
histplot.set_xlabel('Возраст (лет)'); #подписываем абцисс 
histplot.set_ylabel('Количество соискателей'); #подписываем ординат

#получаем логорифмическое значение признака возраст 
log_age= np.log(data_cleaned['Возраст'])
#гистограмма в логарифмическом масштабе
histplot = sns.histplot(log_age , bins=30, ax=axes[1])
#добавялем к графику линии, отображающие среднее и границы интервала метода трех сигм
histplot.axvline(log_age.mean(), color='k', lw=2)
histplot.axvline(log_age.mean()+ 3 * log_age.std(), color='k', ls='--', lw=2)
histplot.axvline(log_age.mean()- 3 * log_age.std(), color='k', ls='--', lw=2)
histplot.set_title('Логарифмическое распредление возраста соискателей'); #заголовок 
histplot.set_xlabel('Возраст в логарифмическом масштабе'); #подписываем абцисс 
histplot.set_ylabel('Количество соискателей'); #подписываем ординат

#вычисляем численный показатель асимметрии 
log_skew = log_age.skew() 
print(f'Численный показатель асимметрии: {log_skew}')

#пишем функцию, которая реализует алгоритм метода z-отклонения 
def outliers_z_score_mod(data, feature, log_scale=False, left=3, right=3):
    
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned

#применяем данную функцию к признаку возраст и выводим количество выбросов 
#включаем логорифмирование и делаем послабление на 1 сигму с правой стороны 
outliers_z, data_cleaned = outliers_z_score_mod(data_cleaned, 'Возраст', log_scale=True, left=3, right=4)
print(f'Число выбросов по методу z-отклонения: {outliers_z.shape[0]}') #выводим количество выбросов 
print(f'Результирующее число записей: {data_cleaned.shape[0]}') #выводим количество результирующих данных 
```

Так как численный показатель ассиметрии положительный, асимметрия распределения является правосторонней. По этой причине мы сделали послабление на 1 сигму в правую сторону
